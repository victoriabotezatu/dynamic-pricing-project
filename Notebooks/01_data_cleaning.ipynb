# Data Cleaning Notebook

The Data Cleaning Notebook documents how the data cleaning process for the dynamic pricing ride-sharing dataset was executed. 

The cleaning process examines:

- dataset structure and data types
- column names
- missing and blank values
- duplicate records
- categorical labels
- invalid numerical values
- minimum, maximum, and potentially unusual numerical values


```python
import pandas as pd 
import os 
```

```python
file_path = "../Data/Raw/dynamic_pricing(in).csv"

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print("Dataset loaded successfully")
else:
    raise FileNotFoundError("The dataset could not be found")
df.head()
```

### Initial Dataset Overview
- The original dataset contains 1,000 observations and 10 variables. Six variables are numerical and four are categorical
- The numerical columns use appropriate integer or floating-point data types

## Column Names 
The main goal of this section is to check if the columns have clear names and do not contain extra spaces 

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
categorical_columns = [
    "Location_Category",
    "Customer_Loyalty_Status", 
    "Time_of_Booking",
    "Vehicle_Type"
]

numerical_columns = [
    "Number_of_Riders", 
    "Number_of_Drivers",
    "Number_of_Past_Rides",
    "Average_Ratings",
    "Expected_Ride_Duration",
    "Historical_Cost_of_Ride"

]
```

```python
def clean_text_values(dataframe,categorical_columns):
    cleaned_dataframe = dataframe.copy()
    for column in categorical_columns:
        cleaned_dataframe[column]=(
            cleaned_dataframe[column].str.strip()
        )
    return cleaned_dataframe
```

```python
df = clean_text_values(df, categorical_columns)
```

Removed leading and trailing spaces from the column names and categorical values in order to prevent identical labels from being treated as different categories because of hidden spaces

## Missing Values 
The main goal of this section is to check if the dataset contains missing values

```python
df.isnull().sum()
```

```python
total_missing_values = df.isnull().sum().sum()
print("Total missing values:", total_missing_values)
```

```python
for column in categorical_columns:
    blank_count=(df[column]=="").sum()
    print(column, "blank values:", blank_count)
```

Conclusion: No missing values or blank categorical values were found, therefore there is no need to impute values or remove observations

## Duplicate Records
The main goal of this section is to check if the dataset contains duplicate rows

```python
duplicate_rows=df.duplicated().sum()
print("Duplicate Rows:", duplicate_rows)
```

Conclusion: No duplicate rows are in the dataset. Therefore, no rows have to be removed 

## Categorical Variables

The main goal of this section is to check the unique values in the categorical columns in order to identify possible spelling mistakes or inconsistent categories

```python
for column in categorical_columns:
    print("\nColumn:", column)
    print("Unique values:", list(df[column].unique()))
    print("Category counts:")
    print(df[column].value_counts().to_string())
```

Conclusion: 
- All categorical variables contain the expected labels, with no spelling inconsistencies or unexpected categories

## Invalid Values

The main goal of this section is to check if the numerical columns contain impossible values (for instance negative numbers or ratings outside the expected range)

```python
numerical_range_summary = pd.DataFrame({
    "Variable": [
        "Number_of_Riders",
        "Number_of_Drivers",
        "Number_of_Past_Rides",
        "Average_Ratings",
        "Expected_Ride_Duration",
        "Historical_Cost_of_Ride"
    ],
    "Minimum": [
        df["Number_of_Riders"].min(),
        df["Number_of_Drivers"].min(),
        df["Number_of_Past_Rides"].min(),
        df["Average_Ratings"].min(),
        df["Expected_Ride_Duration"].min(),
        df["Historical_Cost_of_Ride"].min()
    ],
    "Maximum": [
        df["Number_of_Riders"].max(),
        df["Number_of_Drivers"].max(),
        df["Number_of_Past_Rides"].max(),
        df["Average_Ratings"].max(),
        df["Expected_Ride_Duration"].max(),
        df["Historical_Cost_of_Ride"].max()
    ]
}).round(2)
numerical_range_summary
```

```python
invalid_values = {
    "Non-positive riders":
        (df["Number_of_Riders"] <= 0).sum(),

    "Non-positive drivers":
        (df["Number_of_Drivers"] <= 0).sum(),

    "Negative past rides":
        (df["Number_of_Past_Rides"] < 0).sum(),

    "Ratings below 1":
        (df["Average_Ratings"] < 1).sum(),

    "Ratings above 5":
        (df["Average_Ratings"] > 5).sum(),

    "Non-positive ride duration":
        (df["Expected_Ride_Duration"] <= 0).sum(),

    "Non-positive ride cost":
        (df["Historical_Cost_of_Ride"] <= 0).sum(),
}

invalid_values
```

```python
supply_exceeds_demand = (
    df["Number_of_Drivers"] > df["Number_of_Riders"]).sum()

print(
    "Observations where driver supply exceeds rider demand:",supply_exceeds_demand)
```

Conclusion: 
- Could not find any invalid numerical values
- All counts, ride durations, and historical costs satisfy the defined validity rules
- Assuming a standard 1–5 rating scale (the observed values range from 3.5 to 5.0), all customer ratings are valid
- Driver supply does not exceed rider demand in any observation, but this is more like a descriptive demand-and-supply result rather than a data-validity rule

## Cleaning Summary

The main goal of this section is to create the table summarizing the main data-quality checks previously performed

```python
expected_categories = {
    "Location_Category": ["Urban", "Suburban", "Rural"],
    "Customer_Loyalty_Status": ["Regular", "Silver", "Gold"],
    "Time_of_Booking": ["Morning","Afternoon","Evening","Night"],
    "Vehicle_Type": ["Economy", "Premium"]
}
unexpected_category_count = 0

for column in categorical_columns:
    for value in df[column].unique():
        if value not in expected_categories[column]:
            unexpected_category_count += 1

print("Total unexpected categories:", unexpected_category_count)
```

```python
cleaning_summary = pd.DataFrame({
    "Check": [
        "Rows",
        "Columns",
        "Missing values",
        "Blank categorical values",
        "Duplicate rows",
        "Unexpected categories",
        "Invalid numerical values",
        "Rows removed",
        "Values imputed"
    ],
    "Result": [
        df.shape[0],
        df.shape[1],
        df.isnull().sum().sum(),
        sum(
            (df[column] == "").sum()
            for column in categorical_columns
        ),
        df.duplicated().sum(),
        unexpected_category_count,
        sum(invalid_values.values()),
        0,
        0
    ]
})

cleaning_summary
```

## Save Cleaned Dataset

After we have completed the data-quality checks, the validated dataset is saved for the exploratory data analysis stage

```python
output_folder = "../Data/Cleaned"
cleaned_file_path = (
    "../Data/Cleaned/cleaned_dynamic_pricing.csv"
)

if not os.path.exists(output_folder):
    os.mkdir(output_folder)

df.to_csv(cleaned_file_path, index=False)

if os.path.exists(cleaned_file_path):
    print("Cleaned dataset saved successfully.")
else:
    print("The dataset was not saved.")
```
